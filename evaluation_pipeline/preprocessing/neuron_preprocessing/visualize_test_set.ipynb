```python
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
import pandas as pd
from pathlib import Path
import pickle
```

```python
df_classifier = pd.read_pickle('../../../data/data_tables/df_cell_types.pkl')
```

```python
len(df_classifier)
```

```python
excitatory_cell_types = ['23P', '4P', '5P-IT', '5P-NP', '5P-PT', '6P-IT', '6P-CT']
inhibitory_cell_types = ['MC', 'BC', 'NGC', 'BPC']
```

#### Visualize

```python
sns.catplot(data=df_classifier, x='cell_type_coarse', kind='count')
```

```python
sns.catplot(
    data=df_classifier[
        df_classifier['cell_type'].isin(excitatory_cell_types)
    ].sort_values('cell_type'),
    x='cell_type',
    kind='count',
)
```

```python
sns.catplot(
    data=df_classifier[
        df_classifier['cell_type'].isin(inhibitory_cell_types)
    ].sort_values('cell_type'),
    x='cell_type',
    kind='count',
)
```
