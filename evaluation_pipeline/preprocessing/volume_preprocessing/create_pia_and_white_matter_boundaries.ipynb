```python
# The EM volume is not perfectly aligned and thickness of the cortex varies across the volume.
# We rotate the entire volume 3.5 degrees about the z-axis. We then fit a linear model of the
# pial surface as well as the white matter boundary using a set of manually placed points.
```

```python
import numpy as np
import matplotlib.pyplot as plt
from preprocessing_utils import set_pia_zero
```

```python
# Manually placed points on pia and white matter.
pia = np.load('data/pia_pts.npy')
wm = np.load('data/wm_pts.npy')
```

```python
pia.shape, wm.shape
```

```python
ax = plt.figure().add_subplot(projection='3d')

ax.scatter(*pia[:, [0, 2, 1]].T, label='pia')
ax.scatter(*wm[:, [0, 2, 1]].T, label='wm')

ax.invert_zaxis()

plt.legend()
```

```python
print(f'Pia mean depth: {pia[:, 1].mean()}')
```

```python
print(f'White matter mean depth: {wm[:, 1].mean()}')
```

#### Linear fit of pia and white matter

```python
pia = np.load('data/pia_grid.npy')
wm = np.load('data/wm_grid.npy')
```

```python
ax = plt.figure().add_subplot(projection='3d')

ax.scatter(*pia[:, [0, 2, 1]].T, label='pia')
ax.scatter(*wm[:, [0, 2, 1]].T, label='wm')

ax.invert_zaxis()

plt.legend()
```

```python
print(f'Pia mean depth: {pia[:, 1].mean()}')
```

```python
print(f'White matter mean depth: {wm[:, 1].mean()}')
```

#### Linear fit of pia and white matter with normalized distance

```python
pia_norm = set_pia_zero(pia, pia_grid=pia) / 1000.0
wm_norm = set_pia_zero(wm, pia_grid=pia) / 1000.0
```

```python
ax = plt.figure().add_subplot(projection='3d')

ax.scatter(*pia_norm[:, [0, 2, 1]].T, label='pia')
ax.scatter(*wm_norm[:, [0, 2, 1]].T, label='wm')

ax.invert_zaxis()

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.legend()
```

```python
print(f'Pia mean depth: {pia_norm[:, 1].mean()}')
```

```python
print(f'White matter mean depth: {wm_norm[:, 1].mean()}')
```
