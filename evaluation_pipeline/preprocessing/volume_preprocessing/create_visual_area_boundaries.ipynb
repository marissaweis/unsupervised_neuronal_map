```python
# Rotate volume such that the linear 3D V1-HVA boundary is orthogonal to x-axis for plotting.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from preprocessing_utils import rotate_points, set_pia_zero
```

```python
coords = np.load('data/v1_hva_boundary.npy')
```

```python
ax = plt.figure().add_subplot(projection='3d')

ax.scatter(*coords[:, [0, 2, 1]].T, label='V1-HV1')

ax.invert_zaxis()
ax.set_xlim([0, 1800000])

plt.legend()
```

#### Normalize and rotate to make orthogonal to x-axis

```python
deg = 14
```

```python
pia_grid = np.load('data/pia_grid.npy')
```

```python
area_bounds1 = set_pia_zero(coords, pia_grid=pia_grid) / 1000.0

area_bounds2 = rotate_points(area_bounds1, axis=1, degree=deg)
```

```python
ax = plt.figure().add_subplot(projection='3d')

ax.scatter(*area_bounds1[:, [0, 2, 1]].T, label='V1-HV1')
ax.scatter(*area_bounds2[:, [0, 2, 1]].T, label='V1-HV1_rot')

ax.invert_zaxis()

ax.set_xlim([0, 1800])

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')


ax.view_init(0, -90)

plt.legend()
```
